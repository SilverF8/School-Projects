```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.integrate as scint
import scipy.optimize as opt
from scipy.stats import norm
```

```python
cmeData = pd.read_csv('cme.settle.20221108.s.csv')
# convert columns to date type
cmeData['BizDt'] = pd.to_datetime(cmeData['BizDt']).apply(lambda dt: dt.date())
cmeData['LastTrdDt'] = pd.to_datetime(cmeData['LastTrdDt']).apply(lambda dt: dt.date())
```

```python
def blackScholesForwardPutPremium(F, K, terminalVolatility):
    if terminalVolatility <= 0.0:
        return np.maximum(0.0, K - F)
    d1 = np.log (F / K) / terminalVolatility + 0.5 * terminalVolatility
    d2 = d1 - terminalVolatility
    return K * norm.cdf(-d2) - F * norm.cdf(-d1)
```

```python
def blackScholesTerminalVolatility(F, K, forwardPutPremium):
    if np.isnan(forwardPutPremium):
        return np.nan
    if forwardPutPremium >= K:
        return np.inf
    lowTerminalVolatility = 0.0
    highTerminalVolatility = 10.0
    while (blackScholesForwardPutPremium(F, K, highTerminalVolatility) < forwardPutPremium):
        highTerminalVolatility = 2.0 * highTerminalVolatility
    while (highTerminalVolatility - lowTerminalVolatility > 1.0e-8):
        terminalVolatility = 0.5 * (lowTerminalVolatility + highTerminalVolatility)
        if (blackScholesForwardPutPremium(F, K, terminalVolatility) >= forwardPutPremium):
            highTerminalVolatility = terminalVolatility
        else:
            lowTerminalVolatility = terminalVolatility
    return 0.5 * (lowTerminalVolatility + highTerminalVolatility)
```

```python
futures =\
    cmeData.loc[
        (cmeData['Sym'] == 'BTC') &
        (cmeData['BizDt'] < cmeData['LastTrdDt']) &
        (cmeData['SecTyp'] == 'FUT'),
        ['BizDt', 'LastTrdDt', 'SettlePrice']]
futures['TimeToExpiration'] =\
    (futures['LastTrdDt'] - futures['BizDt']).apply(lambda ts : ts.days / 365.0)
```

```python
options =\
    pd.merge(
        cmeData.loc[
            (cmeData['Sym'] == 'BTC') &
            (cmeData['BizDt'] < cmeData['LastTrdDt']) &
            (cmeData['SecTyp'] == 'OOF') &
            (cmeData['SettlePrice'] != 9_999_999.0) &
            (cmeData['PutCall'] == 0),
            ['BizDt', 'LastTrdDt', 'StrkPx', 'SettlePrice']],
        cmeData.loc[
            (cmeData['Sym'] == 'BTC') &
            (cmeData['BizDt'] < cmeData['LastTrdDt']) &
            (cmeData['SecTyp'] == 'OOF') &
            (cmeData['SettlePrice'] != 9_999_999.0) &
            (cmeData['PutCall'] == 1),
            ['BizDt', 'LastTrdDt', 'StrkPx', 'SettlePrice']],
        on = ['BizDt', 'LastTrdDt', 'StrkPx'],
        suffixes = ('Put', 'Call'))
```

```python
def estimateDiscountFactor(row):
    FK = row.SettlePrice - row.StrkPx
    CP = row.SettlePriceCall - row.SettlePricePut
    return (FK * CP).sum() / (FK ** 2).sum()
futures['DiscountFactor'] =\
    pd.merge(
        futures,
        options.loc[
            (np.isnan(options['SettlePricePut']) == False) &
            (np.isnan(options['SettlePriceCall']) == False)],
        on = ['BizDt', 'LastTrdDt'])\
    .groupby(['BizDt', 'LastTrdDt'])\
    .apply(estimateDiscountFactor)\
    .values
```

```python
def estimateBreakEvenVarianceStrike(row):
    F = np.array(row.SettlePrice)[0]
    T = np.array(row.TimeToExpiration)[0]
    D = np.array(row.DiscountFactor)[0]
    Ks = np.array(row.StrkPx)[1 : -1]
    K0 = np.array(Ks[Ks <= F])[-1]
    dKs = (np.array(row.StrkPx)[2 :] - np.array(row.StrkPx)[: -2]) / 2
    Qs = np.minimum(np.array(row.SettlePricePut)[1 : -1], np.array(row.SettlePriceCall)[1 : -1])
    # 2nd term (optional) due to integration bounds not exactly lining up with futures price
    return (2 * (Qs * dKs / Ks ** 2).sum() / D - (F / K0 - 1) ** 2) / T
futures['BreakEvenVarianceStrike'] =\
    pd.merge(
        futures,
        options.loc[
            (np.isnan(options['SettlePricePut']) == False) &
            (np.isnan(options['SettlePriceCall']) == False)],
        on = ['BizDt', 'LastTrdDt'])\
    .groupby(['BizDt', 'LastTrdDt'])\
    .apply(estimateBreakEvenVarianceStrike)\
    .values
```

```python
def estimateStein2VarianceStrike(kappa, theta, alpha, sigma0, T):
    return\
        theta ** 2 + alpha ** 2 / (2 * kappa) +\
        2 * theta * (sigma0 - theta) * (1 - np.exp(-kappa * T)) / (kappa * T) +\
        ((sigma0 - theta) ** 2 - alpha ** 2 / (2 * kappa)) *\
        (1 - np.exp(-2 * kappa * T)) / (2 * kappa * T)
def stein2ObjectiveFunction(parameters):
    [kappa, theta, alpha, sigma0] = parameters
    penalty = 0.0
    for _, row in futures.iterrows():
        T = row['TimeToExpiration']
        VS0 = row['BreakEvenVarianceStrike']
        error = estimateStein2VarianceStrike(kappa, theta, alpha, sigma0, T) - VS0
        penalty += np.minimum(np.abs(error), 1.0e150) ** 2.0 # prevent floating point overflow
    return penalty
[kappa, theta, alpha, sigma0] =\
    opt.differential_evolution(
        stein2ObjectiveFunction,
        bounds = [(0.0, 10.0), (-10.0, 10.0), (0.0, 10.0), (0.0, 10.0)],
        seed = 0,
        polish = True,
        maxiter = 1000).x
futures['Stein2VarianceStrike'] =\
    np.vectorize(estimateStein2VarianceStrike)(
        kappa, theta, alpha, sigma0, futures['TimeToExpiration'])
```

```python
(fig, ax) = plt.subplots(1, 1, figsize = (8, 5))
fig.suptitle('Bitcoin Options on Futures Break-Even and Fitted Stein & Stein Variance Strikes')
ax.plot(
    futures['TimeToExpiration'],
    futures['BreakEvenVarianceStrike'],
    label = 'Break-Even',
    marker = 'o')
ax.plot(
    futures['TimeToExpiration'],
    futures['Stein2VarianceStrike'],
    label = 'Fitted Stein & Stein')
ax.legend(title = 'Variance Strike')
ax.set_xlabel('Time to Expiration')
ax.set_ylabel('Variance Strike')
ax.table(
    cellText = [[
        r'$\kappa=' + f'{kappa : .7}' + r'$',
        r'$\theta=' + f'{theta : .7}' + r'$',
        r'$\alpha=' + f'{alpha : .7}' + r'$',
        r'$\sigma_0=' + f'{sigma0 : .7}' + r'$']],
    cellLoc = 'center',
    loc = 'top')
ax.table(
    cellText = [
        [f'{T : .3f}' for T in futures['TimeToExpiration']],
        [f'{VS0 : .3f}' for VS0 in futures['BreakEvenVarianceStrike']],
        [f'{vT : .3f}' for vT in futures['Stein2VarianceStrike']]],
    rowLabels = [
        'Time to Expiration',
        'Break-Even\nVariance Strike',
        'Fitted Stein & Stein\nVariance Strike'],
    bbox = [0.1, -0.5, 1, 0.275])
fig.savefig('FinalProject_MSCF_ADM-46915_Fall2022_SolutionChart1.pdf', bbox_inches = 'tight')
```

```python
maximumT = np.array(futures['TimeToExpiration'])[-1]
M = 1_000_000                   # Number of paths
dt = 7 / 365                    # Time step size
N = int(np.ceil(maximumT / dt)) # Maximum number of time steps
dt = maximumT / N
np.random.seed(4)
epsilons = np.random.normal(size = [M, N, 2])
def monteCarloSimulation(rho, F0, T):
    N_ = int(np.ceil(T / dt))
    dt_ = T / N_
    ex = np.exp(-kappa * dt_)
    sigma = np.zeros([M, N_ + 1])
    F = np.zeros([M, N_ + 1])
    sigma[:, 0] = sigma0
    F[:, 0] = F0
    for i in range(N_):
        dW = np.sqrt(dt_) * epsilons[:, i, 0]
        dZ = rho * dW + np.sqrt(np.maximum((1 - rho ** 2) * dt_, 0)) * epsilons[:, i, 1]
        F[:, i + 1] = F[:, i] * np.exp(sigma[:, i] * (dW - 0.5 * sigma[:, i] * dt_))
        sigma[:, i + 1] =\
            theta + ex * (sigma[:, i] - theta) +\
            alpha * np.sqrt((1 - ex ** 2) / (2 * kappa * dt_)) * dZ
    return (F, sigma)
```

```python
combined = pd.merge(futures, options, on = ['BizDt', 'LastTrdDt'])
combined['ForwardPutPremium'] =\
    0.5 * (
        (combined['SettlePricePut'] + combined['SettlePriceCall']) / combined['DiscountFactor'] +
        combined['StrkPx'] - combined['SettlePrice'])
combined['ImpliedVolatility'] =\
    np.vectorize(blackScholesTerminalVolatility)(
        combined['SettlePrice'],
        combined['StrkPx'],
        combined['ForwardPutPremium']) / np.sqrt(combined['TimeToExpiration'])
```

```python
expirationDate = pd.to_datetime('2022-12-30').date()
group = combined[combined['LastTrdDt'] == expirationDate]
F0 = np.array(group['SettlePrice'])[0]
T = np.array(group['TimeToExpiration'])[0]
Ks = np.array(group['StrkPx'])
marketForwardPutPremiums = np.array(group['ForwardPutPremium'])
rhos = np.arange(-1, 1.025, .05)
squareErrors = np.zeros(len(rhos))
for i in range(len(rhos)):
    rho = rhos[i]
    (F, _) = monteCarloSimulation(rho, F0, T)
    FT = F[:, -1]
    for j in range(len(Ks)):
        K = Ks[j]
        marketForwardPutPremium = marketForwardPutPremiums[j]
        modelForwardPutPremium = np.mean(np.maximum(K - FT, 0.0))
        squareErrors[i] += (modelForwardPutPremium - marketForwardPutPremium) ** 2
```

```python
rho = rhos[np.argmin(squareErrors)]
(F, sigma) = monteCarloSimulation(rho, F0, T)
```

```python
(fig, ax) = plt.subplots(1, 1, figsize = (8, 5))
fig.suptitle(f'Bitcoin Options on Futures Stein & Stein Pricing Error\nfor Expiration {expirationDate}')
ax.plot(rhos, squareErrors, label = 'Model Error')
ax.scatter(rho, np.min(squareErrors), label = r'Minimum at $\rho=' + f'{rho : .2f}' + r'$')
ax.legend()
ax.set_xlabel(r'$\rho$')
ax.set_ylabel('Error')
fig.savefig('FinalProject_MSCF_ADM-46915_Fall2022_SolutionChart2.pdf')
```

```python
(fig, ax) = plt.subplots(1, 1, figsize = (8, 5))
fig.suptitle(f'Bitcoin Options on Futures Volatility Smiles\nfor Expiration {expirationDate}')
ax.plot(Ks, group['ImpliedVolatility'], label = 'Market Quotes', marker = 'o')
FT = F[:, -1]
modelVolatilities = []
for K in Ks:
    forwardPutPremium = np.mean(np.maximum(K - FT, 0.0))
    terminalVolatility = blackScholesTerminalVolatility(F0, K, forwardPutPremium)
    modelVolatilities.append(terminalVolatility / np.sqrt(T))
ax.plot(
    Ks, modelVolatilities,
    label = r'Calibrated Stein & Stein Model with $\rho=' + f'{100 * rho : .0f}' + '\%$')
ax.legend()
ax.set_xlabel('Strike Price (log scale)')
ax.set_ylabel('Implied Volatility')
ax.semilogx()
fig.savefig('FinalProject_MSCF_ADM-46915_Fall2022_SolutionChart3.pdf')
```

```python
D = np.array(group['DiscountFactor'])[0]
FT = F[:, -1]
minF = F.min(axis = 1)
Ks_ = np.arange(0, 25_250.0, 500.0)
Ks_[-1] = 25_000.0
Bs = np.arange(0, 18_250.0, 500.0)
Bs[-1] = 18_000.0
(X, Y) = np.meshgrid(Ks_, Bs)
Z = 0.0 * X
for i in range(len(Bs)):
    B = Bs[i]
    for j in range(len(Ks_)):
        K = Ks_[j]
        Z[i, j] = D * np.mean(np.maximum(0, K - FT) * (minF > B))
fig = plt.figure(figsize = (20, 20))
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(X, Y, Z, cmap = 'plasma')
ax.set_title('Down-and-Out Put Premium Surface')
ax.set_xlabel(r'$K$')
ax.set_ylabel(r'$B$')
ax.set_zlabel(r'$P^{\mathrm{down_and_out}}\left(T,K,B\right)$')
ax.view_init(30, 120)
fig.savefig('FinalProject_MSCF_ADM-46915_Fall2022_SolutionChart4.pdf')
```
